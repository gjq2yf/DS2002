```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Top_Highest_Openings.csv')
```

```python
df.head()
```

```python
# line graph to show cumulative total gross earnings over time
# ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# sort by date
df = df.sort_values('Date')

# calculate total grossing
df['Cumulative Total Gross'] = df['Total Gross'].cumsum()

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Cumulative Total Gross'], marker='o', linestyle='-')
plt.title('Cumulative Total Gross Earnings Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Total Gross ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# histogram to show total gross earnings
plt.figure(figsize=(10, 6))
plt.hist(df['Total Gross'], bins=5, color='purple', alpha=0.7)
plt.title('Distribution of Total Gross Earnings')
plt.xlabel('Total Gross Earnings')
plt.ylabel('Frequency')
plt.show()
```

```python
# bar graph to show total gross earnings by distributor
distributor_totals = df.groupby('Distributor')['Total Gross'].sum().reset_index()

plt.figure(figsize=(12, 8))
plt.bar(distributor_totals['Distributor'], distributor_totals['Total Gross'], color='blue')
plt.title('Total Gross Earnings by Distributor')
plt.xlabel('Distributor')
plt.ylabel('Total Gross Earnings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```
